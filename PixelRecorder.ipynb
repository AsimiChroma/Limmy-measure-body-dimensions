```python
import numpy as np
import cv2 as cv
# mouse callback function

l = 337

def store_pixel1(event,x,y,flags,param):
    if event == cv.EVENT_LBUTTONDOWN:
        print('{}, {}'.format(x,y))    
        with open(r"C:\Users\Anqi\Documents\TensorFlow\Train\Labels\Train" + str(l) +".txt", "a") as file_object:
            # Append 'hello' at the end of file
            file_object.write(str(x) + ", " + str(y) + "\n")
        
# Create a black image, a window and bind the function to window
img = cv.imread(r'C:\Users\Anqi\Documents\TensorFlow\Train\Avi\Train' + str(l) +'.jpg')
cv.namedWindow('image')
cv.setMouseCallback('image',store_pixel1)
while(1):
    cv.imshow('image',img)
    if cv.waitKey(20) & 0xFF == 27:
        break
cv.destroyAllWindows()
```

```python
import os
import cv2 as cv
directory = r'C:\Users\Anqi\Documents\TensorFlow\Train\Avi'

j = 288
total_files = 0
for filename in os.listdir(directory):
    total_files = total_files + 1

def store_pixel(event,x,y,flags,param):
    if event == cv.EVENT_LBUTTONDOWN:
        print('{}, {}'.format(x,y))  
        with open(r"C:\Users\Anqi\Documents\TensorFlow\Train\Labels\Train" + str(j) + ".txt", "a") as file_object:
            # Append 'hello' at the end of file
            file_object.write(str(x) + ", " + str(y) + "\n")

while(1):
    if j == total_files+1:
        break
    img = cv.imread(r'C:\Users\Anqi\Documents\TensorFlow\Train\Avi\Train' + str(j) +'.jpg')
    cv.namedWindow('Train' + str(j) +'.jpg')
    cv.setMouseCallback('Train' + str(j) +'.jpg',store_pixel)
    while(1):
        cv.imshow('Train' + str(j) +'.jpg',img)
        if cv.waitKey(20) & 0xFF == 27:
            break
    cv.destroyAllWindows()
    j = j + 1
    
```
